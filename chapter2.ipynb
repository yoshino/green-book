```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
# ポアソン分布からデータを生成する
mu = 3.5 # 平均
n = 50   # データ数は50

data = np.random.poisson(mu, n)
print('len: ', len(data))
print('mean: ', np.mean(data))
print('var: ', np.var(data))
```

```python
# データを描画する
plt.hist(data, range=(0, 9), bins=10)
```
