## 現象がどのような確率分布で説明されそうか？の仮定
統計モデリングで、現象をグラフで表示した後に、どのようなグラフだと仮定することが必要になります。
この本で紹介されている確率分布は以下の５つ紹介されていて、いくつかグラフで図示してみます。
- ポアソン分布
- 二項分布
- 正規分布
- ガンマ分布
- 一様分布

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ポアソン分布
#-------------------
# データが離散値
# ゼロ以上で有限の範囲
# 上限特になし
# 平均 = 分散

from scipy.stats import poisson

ls = [1.0, 3.0, 6.0, 9.0]

for l in ls:
  x = np.arange(0, 9)
  y = poisson.pmf(x, l)
  ys.append(y)
  plt.plot(x,y, label=f'lambda={l}')
plt.legend()

smaples = np.random.poisson(lam=3.0, size=100)
print('平均: ', np.mean(smaples))
print('標準偏差: ', np.mean(smaples))
```

```python
# 二項分布
#-------------------
# データが離散値
# ゼロ以上で有限の範囲(1~N)
# 分散は平均の関数

from scipy.stats import binom

N = 10
ps = [0.1, 0.2, 0.6, 0.8]

for p in ps:
  ks = np.arange(1,11)
  y = binom.pmf(ks, N, p)
  plt.plot(ks,y, label=f'p={p}')
plt.legend()

n = 10
p = 0.2
smaples = np.random.binomial(n, p, size=1000)
mu = np.mean(smaples)
print('平均: ', mu)
print('分散: ', np.var(smaples))
print('分散(nとpから算出): ', (n * p) * (1 - p))
```

```python
# 正規分布
#-------------------
# データが連続値
# 範囲が -∞ ~ +∞
# 分散は平均とは無関係に決まる

from scipy.stats import norm

mu = 10
scales = [1, 2, 4, 16] # 標準偏差

for s in scales:
  range = 25
  xs = np.arange(mu - range, mu + range)
  y = norm.pdf(xs, loc=mu, scale=s)
  plt.plot(xs, y, label=f'scale={s}')
plt.legend()
```
